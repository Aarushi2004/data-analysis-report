```python
import pandas as pd;
```

```python
#load and reading the dataset FOR rfm
df_RFM = pd.read_csv("RFM_Analysis.csv")
df_RFM
```

```python
# Describing the data
df_RFM.describe()
```

```python
# calculating recency score
def R_Score(x):
    if x['Recency'] > 2485 and x['Recency'] <=2526:
        recency = 2
    if x['Recency'] > 2526:
        recency = 3
    else:
        recency = 1
    return recency

df_RFM['R'] = df_RFM.apply(R_Score,axis=1)
```

```python
df_RFM
```

```python
# calculating Frequency score
def F_Score(x):
    if x['Frequency'] > 8 and x['Frequency'] <=28:
        Frequency = 2
    if x['Frequency'] > 28:
        Frequency = 3
    else:
        Frequency = 1
    return Frequency

df_RFM['F'] = df_RFM.apply(F_Score,axis=1)
```

```python
# calculating Monetary Score
def M_Score(x):
    if x['Monetary'] > 8050 and x['Monetary'] <=33683:
        Monetary = 2
    if x['Monetary'] > 33683:
        Monetary = 3
    else:
        Monetary = 1
    return Monetary

df_RFM['M'] = df_RFM.apply(M_Score,axis=1)
```

```python
df_RFM
```

```python
# calculating the combined rfm score
def RFM_Score(x):
    return str(x['R'])+ str(x['F'])+ str(x['M'])
df_RFM['RFM_Score'] = df_RFM.apply(RFM_Score,axis=1)
```

```python
df_RFM
```

```python
# Function to classify each customer based on RFM score
def rfm_segment(row):
    r, f, m = int(row['R']), int(row['F']), int(row['M'])
    score = f"{r}{f}{m}"

    if score == '333':
        return 'Best Customers'
    elif r == 3 and f == 2 and m >= 2:
        return 'Loyal Customers'
    elif r == 3 and f == 1:
        return 'Recent Buyers'
    elif r == 2 and f == 3:
        return 'Frequent Buyers'
    elif r == 1 and f == 3:
        return 'At Risk Loyalists'
    elif score == '111':
        return 'Dormant Customers'
    elif r == 1 and f == 1 and m == 3:
        return 'Big Spenders Gone Silent'
    elif r == 2 and f == 2 and m == 2:
        return 'Potential Loyalists'
    elif r == 3 and f == 3:
        return 'Frequent but low spending'
    else:
        return 'Others'
   
```

```python
df_RFM
```

```python
#Applying the segmentation
df_RFM['Segment'] = df_RFM.apply(rfm_segment, axis=1)

```

```python
#Viewing segment distribution
print(df_RFM['Segment'].value_counts())
```

```python
df_RFM.to_csv('RFM_Customer Segmentation F.csv')
```
